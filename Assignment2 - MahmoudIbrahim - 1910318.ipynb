```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
def linear(x):
    return x
```

```python
x = np.linspace(-10, 10)
```

```python
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function :Linear')
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```

```python
def RELU(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :RELU')
plt.show()
```
